```python
import pandas as pd
import requests
```

```python
key = "otqvshn22qrt"
regionCode = "CL"

url = f"https://api.ebird.org/v2/ref/taxonomy/ebird"

payload={}
headers = {
  'X-eBirdApiToken': key
}
response = requests.request("GET", url, headers=headers, data=payload)
```

```python
text = response.text
```

```python
a = text.split("\n")
headers = a[0].split(",")
body = a[1:]
```

```python
body_aux = []
for i in range(len(body)):
    body[i] = body[i].split(",")
    if len(body[i]) == 15:
        body_aux.append(body[i])
body = body_aux
```

```python
df = pd.DataFrame(body, columns=headers)
```
