```python
def df_clean(df):
    #Hacemos que la columna exoticCategory sea categorica
    df["exoticCategory"] = df["exoticCategory"].astype("category")
    df["exoticCategory"] = df["exoticCategory"].cat.codes
    
    #Observaciones Validas
    # df = df[df["obsValid"] == True]

    #Observaciones Revisadas
    # df = df[df["obsReviewed"] == True]

    #Eliminamos las columnas
    df = df.drop(columns = ["obsReviewed", "obsValid", "errors", "subId"])

    #en la columna howMany reemplazamos lo NaN por 0
    df["howMany"] = df["howMany"].fillna(0)

    #la columna obsDt la convertimos en datetime
    df["obsDt"] = pd.to_datetime(df["obsDt"], errors='coerce')
    df["obsDt"] = df["obsDt"].dt.strftime('%Y-%m-%d 00:00')
    df["obsDt"] = pd.to_datetime(df["obsDt"])

    #Borramos los nan
    df.dropna(inplace =True)

    return df
```

Lo aplicamos por cada año, y guardamos

```python
for year in range(2015,2023):
    df = pd.read_csv(f"../raw/Aves/Avistamientos_y_region/vis_{year}.csv")
    df = df_clean(df)
    #guardamos el archivo
    df.to_csv(f"vis_{year}.csv", index = False)
```
